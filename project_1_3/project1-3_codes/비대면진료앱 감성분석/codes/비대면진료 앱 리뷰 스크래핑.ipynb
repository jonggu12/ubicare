## 라이브러리 불러오기

```python
import requests
import os

from bs4 import BeautifulSoup

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service 

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd 
```

## 구글 앱 스토어 열기

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

url = 'https://play.google.com/store/apps/details?id=com.baedalyakgook_user&hl=ko-KR'  # 닥터나우
url = 'https://play.google.com/store/apps/details?id=com.ksncho.hospitalinfo&hl=ko-KR'  # 굿닥
url = 'https://play.google.com/store/apps/details?id=com.merakiplace.mydoctor&hl=ko-KR' # 나만의닥터

driver.get(url)
```

## 리뷰 Component Page Click

```python
review_page = driver.find_element(By.XPATH, '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[2]/div[2]/div/div[1]/div[1]/c-wiz[4]/section/div/div[2]/div[5]/div/div/button')

review_page.click()
```

## 리뷰 스크래핑

```python
reviews = driver.find_elements(By.CSS_SELECTOR,'div[class="RHo1pe"]')
print(len(reviews))

review_list = []
rating_list = []
for review in reviews:
    review_list.append(review.find_element(By.CSS_SELECTOR, 'div[class="h3YV2d"]').text)
    rating = review.find_element(By.CSS_SELECTOR, 'div[class="iXRFPc"]')
    rating_list.append(rating.get_attribute('aria-label'))

print(review_list)
print(rating_list)
```

## 앱별 데이터프레임 생성 및 dump

> __닥터나우__

```python
doctor_now_data = {
    '리뷰': review_list,
    '평점': rating_list,
}

doctor_now_df = pd.DataFrame(doctor_now_data)
doctor_now_df.index = range(1, len(doctor_now_df) + 1)
doctor_now_df['평점'] = doctor_now_df['평점'].apply(lambda x: int(x[10]))
doctor_now_df.to_csv('닥터나우 앱 리뷰.csv', encoding='utf-8-sig', index=False)
doctor_now_df
```

> __굿닥__

```python
good_doc_data = {
    '리뷰': review_list,
    '평점': rating_list,
}

good_doc_df = pd.DataFrame(good_doc_data)
good_doc_df.index = range(1, len(good_doc_df) + 1)
good_doc_df['평점'] = good_doc_df['평점'].apply(lambda x: int(x[10]))
good_doc_df.to_csv('굿닥터 앱 리뷰.csv', encoding='utf-8-sig', index=False)
good_doc_df
```
